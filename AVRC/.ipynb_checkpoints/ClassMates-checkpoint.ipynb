```python
"""
Create and visualize the network 
"""
import networkx as nx
import matplotlib.pyplot as plt
import dzcnapy_plotlib as dzcnapy
import csv
import pandas as pd
import io
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
df = pd.read_csv("classmates_AdjMatrix.csv")
df 

G = nx.DiGraph()
for col in df:
    for x in list(df.loc[df[col] == 1,'Name']):
        G.add_edge(col,x,weight = df.loc[df[col]])
for col in df:
    for x in list(df.loc[df[col] == 2,'Name']):
        G.add_edge(col,x,weight = df.loc[df[col]])

        
"""      
with open("email-Eu-core.csv") as infile:
    csv_reader = csv.reader(infile)
    G = nx.Graph(csv_reader)
print(G.nodes())
"""

```

```python
# Draw four layouts in four subplots
_, plot = plt.subplots(2, 1)

subplots = plot.reshape(1, 2)[0]
layouts = (nx.random_layout, nx.circular_layout)
titles = ("Random", "Circular")
for plot, layout, title in zip(subplots, layouts, titles):
    pos = layout(G)
    nx.draw_networkx(G, pos=pos, ax=plot, with_labels=False, 
                     **dzcnapy.medium_attrs)
    plot.set_title(title)
    dzcnapy.set_extent(pos, plot)
```

```python
def slice_network(G, T, copy=True):
    """
    Remove all edges with weight<T from G or its copy.
    """
    F = G.copy() if copy else G
    F.remove_edges_from((n1, n2) for n1, n2, w in list(F.edges(data="weight"))
                        if w < T)
    return F
```

```python
for i, threshold in enumerate([0, 0.7, 0.8, 0.9, 0.95, 1.0]):
    ax = plt.subplot(3, 2, i + 1)
    slice_network(G, threshold, copy=False)
    nx.draw_networkx_edges(G, pos, alpha=0.65, 
                           ax=ax, **dzcnapy.small_attrs)
    nx.draw_networkx_nodes(G, pos, ax=ax, **dzcnapy.small_attrs)
    dzcnapy.set_extent(pos, ax, "Threshold={}".format(threshold))

dzcnapy.plot("slicing")
```
